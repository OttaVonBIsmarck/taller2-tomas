# Pandas para limpieza de datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("BikePrices.csv")
```

```python
df.head()
```

### Datos faltantes

```python
df.dropna(axis='columns')
```

```python
df['Model'].value_counts()
```

```python
df['Owner'].value_counts()
```

```python
fig, ax = plt.subplots()

colores = ['tab:red', 'tab:blue', 'tab:orange', 'tab:red']

ax.bar(df["Owner"].unique(), df['Owner'].value_counts(), color=colores)

ax.set_ylabel('Nivel de inventario')
ax.set_title('Inventario de productos')

plt.show()
```

En esta gráfica se muestra como se distribuye el numero de dueños que tienen la bicicletas en la serie de datos. Se muestra que la mayoria de la biciletas dson e unico dueño.

```python
import seaborn as sns
sns.histplot(data=df, x="KM_Driven")
```

Este histograma de los kilometros andados muestra que la mayhoria de bicicletas no ha  isdo andadas mas de 100.000km, mientras que se presentan varios datos atípicos bastante superiores a la media.

```python
sns.violinplot(data=df, x="Seller_Type", y="Year")
```

Esta gráfica de violin compara el tipo de vendedor contra el año en el que se vendio. Se muestra que la media de años no es significativa entre individuos o Dealers. Sin embargo, se muestra que la mayoria de ventas individuales se concentran entre 2010 y 2020, mientas que la venta de Dealers han vendido mayor cantidad antes del 2005 que los individuos.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.scatter('Year', 'Selling_Price', data=df)
ax.set(xlabel='Año', ylabel='Precio de Venta')
plt.show()
```

En esta gráfica de puntos se evidencia una clara tendencia a que se vendan bicicletas mas nuevas que viejas. A lo sumo, la mayoria de punto se encuentran alrededor del año 2010 y 2020.

```python
x = df["Selling_Price"]
y = df["Ex_Showroom_Price"]

plt.scatter(x, y)
plt.scatter(x, x)
plt.show()
```

El gráfic de dispersión evidencia que el precio de exhibición y el precio de venta tienen una correlación positiva, por lo que muestra una tendencia clara en ambos precios
